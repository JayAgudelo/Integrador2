```python
%load_ext autoreload
%autoreload 2

import sys
import os

# Agregar la carpeta raíz del proyecto al path
sys.path.append(os.path.dirname(os.getcwd()))
```

```python
from helpers.fetcher import get_spotify_dataset

df = get_spotify_dataset()
df.head()
```

```python
#drop rows with <15 popularity
df = df[df['popularity'] > 15]
```

```python
#Train split-antes de el preprocesamiento
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['genre'])
print(f"Train set: {train_df.shape}, Test set: {test_df.shape}")
```

```python
from helpers.preprocessor import DataPreprocessor
preprocessor = DataPreprocessor(train_df)
```

```python
preprocessor.drop_nulls()
preprocessor.remove_duplicates()
preprocessor.drop_columns(['Unnamed: 0','artist_name','track_name','track_id','year'])
```

```python
preprocessor.filter_top_genres(top_n=50)
```

```python
preprocessor.get_dataframe()
```

```python
preprocessor.log_transform()
```

```python
preprocessor.transform(keep_original_names=True)
```

```python
train_processed = preprocessor.get_dataframe()
train_processed.head()
```

```python
test_processed = preprocessor.apply_to_test(test_df, keep_original_names=True)
```

```python
test_processed.head()
```

```python
train_processed['popularity'] = train_df.loc[train_processed.index, 'popularity']
test_processed['popularity'] = test_df.loc[test_processed.index, 'popularity']
```

```python
preprocessor.save_preprocessor('../models/preprocessor.joblib')
```

```python
train_processed.to_csv('../temp/train_processed.csv', index=False)
test_processed.to_csv('../temp/test_processed.csv', index=False)

print(f"\n✓ Datasets guardados:")
print(f"  Train: ../temp/train_processed.csv ({train_processed.shape})")
print(f"  Test: ../temp/test_processed.csv ({test_processed.shape})")
```
